#Linear and Multiple Linear Regression

Linear regression (One independent variable to make a prediction)

Multiple linear regresion (two or more independent variables to make a prediction)

Keywords:
Predictor (Indepndent variable), target (dependent variable).

y = b0 + b1(x)

With this relationship we assume a linear prediction.

With training points we fit the model, the results of the trainned points are the parameters (b0 and b1).

We have also noise, because there are other factors what affect the target.

Dont forget that the model isn't always correct, only make a prediction.

1. Import linear model

#Linear regression and multiple:

```python
#Librarie:

from sklearn.linear_model import LinearRegression

#Create the object:
lm =LinearRegression()
lm
```

```python
#Set the predictor variable and target:
X = df[['highway-mpg']]
Y = df['price']

lm.fit(X,Y)
```

```python
Yhat=lm.predict(X)
Yhat[0:5]   
```

```python
lm.intercept_
```
