#Importing and Exporting Data in Python:



```python
import pandas as pd

#Importing a csv wihtout a header:
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
df = pd.read_csv(url, header = None) 
df.head(12)
df.tail(12)
```

```python
#Replace default header
headers = ["symboling", "Normalizaed-losses", 'make', 'fuel-type', 'aspiration', 'num-of-doors',
           'body-style', 'drive-wheels', 'engine-location','wheel-base', 'length', 'width', 'heigth','curb-weight', 'engine-type', 'num-of-cylinders', 'engine-size','fuel-system', 'bore', 'stroke','compression ratio', 'horsepower', 'peak-rpm',
           'city-mpg','highway-mpg', 'price']
df.columns = headers 

path = '/content/data.csv'
df.to_csv(path)
```
