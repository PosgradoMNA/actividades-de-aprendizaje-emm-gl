```python
from typing import Text
##### Quiz #####

class analysedText(object):
    
    def __init__ (self, text):
        # TODO: Remove the punctuation from <text> and make it lower case.
        # TODO: Assign the formatted text to a new attribute called "fmtText"
        self.fmtText = str(text)
        self.fmtText.replace(".", "")
        self.fmtText.replace("!", "")
        self.fmtText.replace(",", "")
        self.fmtText.replace("?", "")
        self.fmtText.lower()
        
    def freqAll(self): 
      self.Dictionary = {} 
      self.listValues = []
      
      self.SetText = set(self.fmtText.Split())          #Get Keys
      self.listKeys = list(self.SetText)

      for word in self.SetText:                         #Get Values
        self.listValues.append(self.SetText.count(word))

      for i in self.listKeys:
        Dictionary = dict.fromkeys(self.listKeys[i], self.listValues[i])
      
      return Dictionary

    def freqOf(self, word):
      # TODO: return the number of occurrences of <word> in <fmtText>
       self.freq = self.fmtText.count(word)
       return self.freq
      
```
